```python
from sklearn.neighbors import KDTree
from biopandas.pdb import PandasPdb
import pandas as pd

def grid(n,r):
    center = [0,0,0]
    R = []
    for l in range(n):
        for m in [0,1,2,10,11]:
            R.append([r*np.cos(2*np.pi*m/n)*np.cos(2*np.pi*l/n),
                     r*np.cos(2*np.pi*m/n)*np.sin(2*np.pi*l/n),
                     r*np.sin(2*np.pi*m/n)])
    R.extend([[0,0,r],[0,0,-r]])
    return np.array(R)#top down is overlapped

def phase_control(zahyo,index,r,id_list,t):
    for i_num,id in enumerate(index):
        tree = KDTree(zahyo, leaf_size=2)
        ind = tree.query_radius(zahyo, r=r[i_num])
        for pair in ind[i_num*t:id]:
            for point in pair[pair >= id]:#and pair < i_num*t
                id_list[point] = 0
    return np.array(id_list)  

def itp_to_pdb(keyword):
    with open(itp_path + keyword + '.itp','r') as f:
      memo_lines = []
      flag = 0
      for line in f:    
        if '[ atoms ]' in line:
          flag = 1
          continue
        elif '[ position_restraints ]' in line:
          flag = 0
          break
        if flag:
          memo_lines.append(line.rstrip('\n'))
      #memo_linesの余計なものを除去
      memo_beads_type = []
      memo_charge = []
      for line in memo_lines[:-1]:
        if line[4:5] == 'S':
          memo_beads_type.append('SS')
          memo_charge.append(float(line[-4:-1]))
        elif line[4:5] == 'T':
          memo_beads_type.append('TT')
          memo_charge.append(float(line[-4:-1]))
        elif line == '':
          pass
        else:
          memo_beads_type.append('RR')
          memo_charge.append(float(line[-4:-1]))
    pdb = PandasPdb().read_pdb(data_path+keyword[:-1]+'.pdb')
    return pd.Series(memo_beads_type).values, memo_charge

def main(protein_num, keyword,n):
    dic_r = {'RR':2.35,'SS':2.05,'TT':1.7}
    path = data_path + keyword[:-1] + '.pdb'
    zahyo = PandasPdb().read_pdb(path).df['ATOM'][['x_coord','y_coord','z_coord']].values
    zahyo_rev = []
    zahyo_index = []
    r_list = []
    type_list,charge_list = itp_to_pdb(keyword)
    for t_num, type in enumerate(type_list):#RR:,SS:,TT:
        zahyo_rev.extend(grid(n,dic_r[type]) + zahyo[t_num])
        zahyo_index.append(len(grid(n,dic_r[type]))*(t_num + 1))
        r_list.append(dic_r[type])
    id_list = [1 for _ in range(len(zahyo_rev))]
    t = phase_control(zahyo_rev,zahyo_index,r_list,id_list,len(grid(n,dic_r[type])))
    #print(len(t[t==1]))#asa atom
    #print(charge_list)#charge
    save_id = []
    for id, marker in enumerate(t):
        if marker == 1:
            save_id.append(id)
    save_charge = []
    for id in save_id:
        save_charge.append(charge_list[id//len(grid(n,dic_r[type]))])
    #print(save_id)
    #print(save_charge)
    #1,search within 2A
    zahyo_point = [zahyo_rev[q_id] for q_id in save_id]
    tree = KDTree(zahyo_point, leaf_size=2)
    ind_point = tree.query_radius(zahyo_point, r=2.0)
    patchiness = 0
    check_point = 0
    for p_num, pair in enumerate(ind_point):
        #check_point += len(pair)
        for pair_index in pair:
            if p_num < pair_index:
                check_point += 1
                if save_charge[p_num] == 1 and save_charge[pair_index] == 1:
                    patchiness += 1
                elif save_charge[p_num] == -1 and save_charge[pair_index] == -1:
                    patchiness += 1
                elif save_charge[p_num] == 0 or save_charge[pair_index] == 0:
                    patchiness += 0
                else:
                    patchiness -= 1
    #print(patchiness/check_point)
    with open(research_path + 'check'+ keyword +'_patchiness.csv', 'a') as f:
        writer = csv.writer(f,delimiter=',')
        writer.writerow(['sample', 'div', 'patchiness'])
        writer.writerow([keyword,n,patchiness/check_point])    
```

```python
def make_dataframe_patchiness(n,keyword):
    i_path = research_path + 'check' + keyword + '_patchiness.csv'
    memo_list = []
    with open(i_path,'r') as f:
        reader = csv.reader(f)
        for fnum, point in enumerate(reader):
            if fnum % 2 != 0:
                memo_list.append(point)
    return memo_list
    
cmap = plt.get_cmap("jet")
height = []
data = []
colors = []

for fname in ["sample41","sample51","sample61","sample71","sample81","sample91"]:
    data.extend(make_dataframe_patchiness(12,fname))
    
for index, point in enumerate(data):
    height.append(float(point[2]))
    colors.append(cmap(index/6))
data = []    
   
for fname in ["minus16","plus23","plus217","plus210","plus213"]:
    data.extend(make_dataframe_patchiness(12,fname))

for index, point in enumerate(data):
    height.append(float(point[2])) 
print(height)

colors.extend(["pink","cyan",cmap(0/4),cmap(2/4),cmap(3/4)])

name = ["sample4","sample5","sample6","sample7","sample8","sample9","-4","+4","+4 -> -4","+4 n<->c","+5"]
                  
ax1 = plt.subplot()
x = np.arange(1,12)
ax1.bar(x,height,width=0.22,color=colors, hatch='/',alpha=0.5)
ax1.set_xticks(x,name)
ax1.set_xlabel('sample')
ax1.set_ylabel('patchiness')
ax1.set_title('sample vs patchiness')
ax1.grid()
plt.xticks(rotation=45)
plt.rcParams["figure.subplot.bottom"] = 0.15
plt.savefig(pic_path + 'all_patchiness',transparent=True)
plt.show()
```
